# Spurious Correlation: Las Vegas Weddings vs. Polar Bear Activity
Do weddings in Las Vegas activate polar bears? A monthly seasonal pattern analysis showing how shared seasonality creates spurious correlations.

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import numpy as np
```

## Define the Dataset
Monthly data: Las Vegas weddings and polar bear active hours (San Diego Zoo) over an average year.

```python
data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Vegas_Weddings": [5100, 6200, 5800, 6900, 7400, 8100, 8500, 7900, 7100, 6800, 5900, 6400],
    "Polar_Bear_Active_Hours": [2.1, 2.4, 2.8, 3.5, 4.1, 4.8, 5.2, 4.9, 3.9, 3.2, 2.5, 2.2],
}

df = pd.DataFrame(data)
df
```

## Explore Data

```python
print(df.describe())
print("\n")
df.info()
```

## Calculate Correlation Coefficient

```python
correlation = df["Vegas_Weddings"].corr(df["Polar_Bear_Active_Hours"])
print(f"The correlation coefficient is: {correlation:.4f}")
```

## Dual Y-Axis Visualization
Las Vegas weddings (pink, left axis) vs. polar bear active hours (cyan, right axis).

```python
fig, ax1 = plt.subplots(figsize=(10, 5))

# Left axis: Vegas weddings
color = "hotpink"
ax1.set_xlabel("Month", fontsize=12)
ax1.set_ylabel("Number of weddings in Las Vegas", color=color, fontsize=12)
ax1.plot(df["Month"], df["Vegas_Weddings"], color=color, marker="o", lw=2.5, ms=10)
ax1.tick_params(axis="y", labelcolor=color)
ax1.grid(True, linestyle="--", alpha=0.3)

# Right axis: Polar bear activity
ax2 = ax1.twinx()
color = "tab:cyan"
ax2.set_ylabel("Polar bear active hours (San Diego Zoo)", color=color, fontsize=12)
ax2.plot(df["Month"], df["Polar_Bear_Active_Hours"], color=color, marker="o", lw=2.5)
ax2.tick_params(axis="y", labelcolor=color)

plt.title(
    f"Do Vegas Weddings Activate Polar Bears?\nLas Vegas Weddings vs. Polar Bear Activity (r = {correlation:.2f})",
    fontsize=14,
    fontweight="bold",
    pad=15,
)
fig.tight_layout()
plt.show()
```

## Statistical Significance Test

```python
r, p_value = stats.pearsonr(df["Vegas_Weddings"], df["Polar_Bear_Active_Hours"])
print(f"Pearson correlation: r = {r:.4f}")
print(f"P-value: p = {p_value:.6f}")
print(f"\nStatistically significant (p < 0.05)? {'Yes' if p_value < 0.05 else 'No'}")
print(f"\nR² = {r**2:.4f} — weddings 'explain' {r**2*100:.1f}% of the variation in polar bear activity")
print(f"\nOf course, getting married in Vegas does NOT make polar bears more active!")
```

## Scatter Plot with Regression Line

```python
fig, ax = plt.subplots(figsize=(8, 5))

ax.scatter(df["Vegas_Weddings"], df["Polar_Bear_Active_Hours"],
           color="mediumpurple", s=120, zorder=5, edgecolors="black")

# Regression line
slope, intercept, r_value, p_val, std_err = stats.linregress(
    df["Vegas_Weddings"], df["Polar_Bear_Active_Hours"]
)
x_line = np.linspace(df["Vegas_Weddings"].min(), df["Vegas_Weddings"].max(), 100)
y_line = slope * x_line + intercept
ax.plot(x_line, y_line, color="red", linewidth=2, linestyle="--",
        label=f"Regression line (R²={r_value**2:.2f})")

# Month labels
for i, txt in enumerate(df["Month"]):
    ax.annotate(txt, (df["Vegas_Weddings"][i], df["Polar_Bear_Active_Hours"][i]),
                xytext=(5, 5), textcoords="offset points", fontsize=9)

ax.set_xlabel("Number of Las Vegas Weddings", fontsize=12)
ax.set_ylabel("Polar Bear Active Hours", fontsize=12)
ax.set_title("Scatter Plot: Vegas Weddings vs. Polar Bear Activity", fontsize=13, fontweight="bold")
ax.legend(fontsize=11)
ax.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("THE HIDDEN CONFOUNDER: SEASONALITY")
print("="*60)
print("""
Both variables follow the SAME seasonal pattern:
- Summer: more weddings (vacation season) AND more polar bear 
  activity (longer daylight hours)
- Winter: fewer weddings AND less polar bear activity

The confounder here is TIME OF YEAR / SEASON:
- Warm weather → more tourists → more Vegas weddings
- Warm weather → longer days → more active zoo animals

This is a classic example of how seasonal patterns create 
spurious correlations between completely unrelated phenomena.

Lesson: Always check for seasonal/cyclical confounders!
""")
```
